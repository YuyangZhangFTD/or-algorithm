```python
import pandas as pd
from copy import deepcopy
from matplotlib import pyplot as plt
```

```python
# number of nodes
# 0 for origin node
# n <= 1000
number_nodes = 100
```

```python
# read data
dt = pd.read_csv("input/input_distance-time.txt")
from_to_node = list(zip(dt["from_node"].values, dt["to_node"].values))
ds = pd.Series(dt["distance"].values, index=from_to_node)
del dt, from_to_node

node = pd.read_csv("input/input_node.csv")
node1 = node[node.type == 1]    # origin node
node2 = node[node.type == 2]
node2 = node2.loc[:, ["ID", "lng", "lat"]].sample(n=number_nodes)

nid_list = node2["ID"].values.tolist()
```

```python
# init
route_list = [
    ([0, nid, 0], ds[0, nid] + ds[nid, 0])
    for nid in nid_list
]
route_list
```

```python
def tsp_cw(route_list, ds):
    route_list = deepcopy(route_list)
    while len(route_list) > 1:
        saving_value_list = []
        for r_i in range(len(route_list)):
            for r_j in range(r_i+1, len(route_list)):
                n_i = route_list[r_i][0][-2]      # (0, ..., i, 0)
                n_j = route_list[r_j][0][1]      # (0, j, ..., 0)
                sv = ds[(n_i, 0)] + ds[(0, n_j)] - ds[(n_i, n_j)] 
                saving_value_list.append((r_i, r_j, sv))
        saving_value_list.sort(key=lambda x: x[-1], reverse=True)
        r_i, r_j, sv = saving_value_list[0]
        if sv < 0:
            return route_list
        else:
            route_list[r_i] = [
                route_list[r_i][0][:-1] + route_list[r_j][0][1:], 
                route_list[r_i][1] + route_list[r_j][1] - sv
            ]
            del route_list[r_j]
    return route_list
            
```

```python
final_route = tsp_cw(route_list, ds)
final_route
```
